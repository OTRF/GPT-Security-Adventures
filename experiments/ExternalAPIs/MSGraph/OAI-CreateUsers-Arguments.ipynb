## Download MS Graph (User) OpenAPI specification

Location: https://github.com/microsoftgraph/msgraph-sdk-powershell/blob/dev/openApiDocs/v1.0/Users.yml

## Load Open API Specification

```python
from langchain.chains.openai_functions.openapi import openapi_spec_to_openai_fn
from langchain.utilities.openapi import OpenAPISpec
import os

current_directory = os.path.abspath('')
data_directory = os.path.join(current_directory, "data")
msgraph_api_file = os.path.join(data_directory, "OpenAPI-MSGraph-v1.0-Users.yml")

openapi_spec = OpenAPISpec.from_file(msgraph_api_file)
```

```python
graph_openai_functions, graph_callables = openapi_spec_to_openai_fn(openapi_spec)
```

```python
graph_openai_functions[1]
```

```python
# InvalidRequestError: '$.functions' is too long. Maximum length is 128, but got 179 items.
createuser_graph_openai_function = graph_openai_functions[1]
```

## Initialize Open AI Key

```python
from langchain.chat_models import ChatOpenAI
import os
from dotenv import load_dotenv
import openai

# Get your key: https://platform.openai.com/account/api-keys
load_dotenv()
openai.api_key = os.getenv("OPENAI_API_KEY")
```

## Initialize ChatOpenAI Model

```python
model = ChatOpenAI(temperature=0).bind(functions=[createuser_graph_openai_function])
```

## Generate Function Arguments

```python
model.invoke("Create user Roberto Rodriguez in the security research department working in the US and as a disabled account. ")
```
